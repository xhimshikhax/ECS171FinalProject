```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import operator

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
# load data 
cars = pd.read_csv('./project.csv')
# mpg, cylinders, displacement, horsepower, weight, acceleration, model year, origin, car name
```

```python
sns.set_style("whitegrid");
sns.pairplot(cars, hue="mpg", height=3); 
plt.show()
```

```python
df = pd.DataFrame(cars)  
sns.heatmap(cars.corr(), annot = True, vmin=-1, vmax=1, center= 0) 
```

Based on the heatmap we can see that there is a stronger relationship between mpg and the attributes cylinders, displacement, horsepower and weight. 

We started off by building a neural network that trained an all of the attributes

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler

# split data into inputs & output
X = cars.drop(['mpg','car name'], axis = 1)
y = cars['mpg']

# normalize data
scaler = MinMaxScaler(feature_range=(0,1))
X_rescaled = scaler.fit_transform(X)
X = pd.DataFrame(data = X_rescaled, columns = X.columns)

# split data into train & test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

#Learning rate should be 0.3
#Number of epochs should be 500
mlp = MLPRegressor(solver='adam', random_state = 42, activation='relu', learning_rate_init=0.3, batch_size = 100, hidden_layer_sizes=(4,), max_iter = 500)

mlp.fit(X_train, y_train)
score = mlp.score(X_test, y_test)
pred = mlp.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = mlp.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)
R2 = r2_score(y_test, y_pred)

print("MSE: ", MSE)
print("R2: ", R2)
```

After looking at the output of the above neural network, we decided to them run one where we only have the attributes that seem to have the most correlation to mpg. This is determined to be cylinders, displacement, horsepower and weight by our heatmap. 

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler

# split data into inputs & output
X = cars.drop(['mpg', 'car name', 'acceleration', 'model year', 'origin'], axis = 1)
y = cars['mpg']

# normalize data
scaler = MinMaxScaler(feature_range=(0,1))
X_rescaled = scaler.fit_transform(X)
X = pd.DataFrame(data = X_rescaled, columns = X.columns)

# split data into train & test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

#Learning rate should be 0.3
#Number of epochs should be 500
mlp = MLPRegressor(solver='adam', random_state = 42, activation='relu', learning_rate_init=0.3, batch_size = 100, hidden_layer_sizes=(4,), max_iter = 500)

mlp.fit(X_train, y_train)
score = mlp.score(X_test, y_test)
pred = mlp.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = mlp.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)
R2 = r2_score(y_test, y_pred)

print("MSE: ", MSE)
print("R2: ", R2)
```

```python
import operator

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures

x = cars['cylinders']
y = cars['mpg']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)

for i in range(1,5):

    polynomial_features= PolynomialFeatures(degree = i, include_bias=False)

    x_poly_train = polynomial_features.fit_transform(np.asarray(x_train).reshape(-1,1))
    x_poly_test = polynomial_features.fit_transform(np.asarray(x_test).reshape(-1,1))

    model = LinearRegression()
    model.fit(x_poly_train,np.asarray(y_train).reshape(-1,1))
   
    y_train_pred = model.predict(x_poly_train)
    y_test_pred = model.predict(x_poly_test)


    plt.scatter(x, y, s=10)
    # sort the values of x before line plot
    sort_axis = operator.itemgetter(0)
    sorted_zip = sorted(zip(x_train,y_train_pred), key=sort_axis)
    x_axis, y_axis = zip(*sorted_zip)
    plt.plot(x_axis, y_axis, color='m')
    plt.show()
```

### displacement (degree 2) ###

```python
import operator

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures

x = cars['displacement']
y = cars['mpg']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)

for i in range(1,9):

    polynomial_features= PolynomialFeatures(degree = i, include_bias=False)

    x_poly_train = polynomial_features.fit_transform(np.asarray(x_train).reshape(-1,1))
    x_poly_test = polynomial_features.fit_transform(np.asarray(x_test).reshape(-1,1))

    model = LinearRegression()
    model.fit(x_poly_train,np.asarray(y_train).reshape(-1,1))
   
    y_train_pred = model.predict(x_poly_train)
    y_test_pred = model.predict(x_poly_test)
    
    MSE = mean_squared_error(y_test, y_test_pred)
    RMSE = np.sqrt(MSE)
    R2 = r2_score(y_test, y_test_pred)
    print("MSE:",MSE)
    print("RMSE:",RMSE)
    print("R2:\n",R2)

    print("This is plot in degree " + str(i) )
    plt.scatter(x, y, s=10)
    # sort the values of x before line plot
    sort_axis = operator.itemgetter(0)
    sorted_zip = sorted(zip(x_train,y_train_pred), key=sort_axis)
    x_axis, y_axis = zip(*sorted_zip)
    plt.plot(x_axis, y_axis, color='m')
    plt.show()
```

## Situation that only use four high related attributes ##

```python
train, test = train_test_split(cars, test_size=0.2, random_state = 21)
X_train, y_train = train.drop(columns=['mpg', 'car name', 'acceleration', 'model year', 'origin']), train['mpg']
X_test, y_test = test.drop(columns=['mpg', 'car name', 'acceleration', 'model year', 'origin']), test['mpg']
```

```python

lr = LinearRegression()

lr.fit(X_train, y_train)

predicted = lr.predict(X_test)

mean_squared_error(predicted, np.asarray(y_test).reshape(-1, 1))
```

```python
for degree in range(1, 8):
    poly = PolynomialFeatures(degree=degree)
    x_poly_train = poly.fit_transform(X_train)
    x_poly_test = poly.fit_transform(X_test)
    model = LinearRegression()
    model.fit(x_poly_train, y_train)

    y_train_pred = model.predict(x_poly_train)
    y_test_pred = model.predict(x_poly_test)

    # Calculate the RMSE and R2 for the training set

    rmse = np.sqrt(mean_squared_error(y_train,y_train_pred))
    r2 = r2_score(y_train,y_train_pred)
    print('\nFor degree: '+str(degree))
    print('\nTraining RMSE: %8.15f' % rmse)
    print('Training R2: %8.15f' % r2)

    # Reporting for the test set:
    rmse_test = np.sqrt(mean_squared_error(y_test,y_test_pred))
    r2_test = r2_score(y_test,y_test_pred)
    print('\nTest RMSE: %8.15f' % rmse_test)
    print('Test R2: %8.15f' % r2_test)
#plr.fit(X_poly_test, np.asarray(y_test).reshape(-1, 1))
    
#predicted = plr.predict(poly.transform(np.asarray(X_test).reshape(-1, 1)))
#test = mean_squared_error(predicted, np.asarray(y_test).reshape(-1, 1))
#print("testing MSE error",test)

#plr.fit(X_poly_train, np.asarray(y_train).reshape(-1, 1))

#predicted = plr.predict(poly.transform(np.asarray(X_train).reshape(-1, 1)))
#train = mean_squared_error(predicted, np.asarray(y_train).reshape(-1, 1))
#print("train MSE error",train)

#print("degree2:",plr.intercept_, plr.coef_)
```
